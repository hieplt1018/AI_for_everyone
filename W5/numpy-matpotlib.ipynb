## Matplotlib

Matploblib là một thư viện trực quan hoá dữ liệu phổ biến trong Python. Nó có thể vẽ được nhiều loại đồ thị khác nhau, và rất hữu ích khi làm việc cùng với NumPy.

## Numpy

Một mảng numpy là một tập hợp các giá trị cùng kiểu dữ liệu và được đánh số bằng các số nguyên dương. Numpy là module quan trọng cho việc sử lý dữ liệu và có thể chuyển đổi qua kiểu dữ liệu tensor trong tensorflow và pytorch.

Ở bài viết này, chúng ta sẽ sử dụng mảng numpy để tính toán và làm input cho các đồ thị matplotlib một số hàm cơ bản.

### VD1: Vẽ đồ thị đường thẳng 

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(8, 10)  #tạo mảng numpy từ [8,10]
y = 8*x + 10
plt.title("Đồ thị y = 8*x + 10") #Đặt title cho figure
plt.xlabel("Trục x") #Đặt title cho trục x
plt.ylabel("Trục y") #Đặt title cho trục y
plt.plot(x,y) #Vẽ đồ thị với trục x so với trục y
plt.show() #Hiển thị figure
```

### VD2: Vẽ đồ thị hàm sphere (đồ thị phương trình bậc 2)

```python
import numpy as np
from matplotlib import pyplot as plt

#np.arange(số đầu, số cuối, khoảng cách giữa các giá trị) 
x = np.arange (-10,10,0.8) 
y = x*x
plt.title("Đồ thị y = x^2")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.plot(x,y)
plt.show()
```

### VD3: Vẽ đồ thị hàm cosine

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(-10, 10, 0.01)
y = np.cos(x)
plt.title("Đồ thị hàm y = cos(x)")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.plot(x,y)
plt.show()
```

### VD4: Vẽ đồ thị hàm tanh

```python
import numpy as np
from matplotlib import pyplot as plt

def tanh(x):
    ex = np.exp(x)
    emx = np.exp(-x)
    return (ex - emx) / (ex + emx)

x = np.arange(-5, 5, 0.1)
a = tanh(x)
dx = 1 - a**2 #Tính đạo hàm của tanh(x)

fig, ax = plt.subplots(figsize=(10, 5)) #Tạo 2 axes cùng 1 figure
ax.spines['left'].set_position('center') #chuyển trục bên trái về trung tâm figure
ax.spines['bottom'].set_position('center') #chuyển trục dưới về trung tâm figure
ax.spines['right'].set_color('none') #xóa trục phải
ax.spines['top'].set_color('none') #xóa trục trên
ax.xaxis.set_ticks_position('bottom') #thiết lập trục x là trục dưới
ax.yaxis.set_ticks_position('left') #thiết lập trục y là trục trái

ax.plot(x, a, color="#d35400", linewidth=3, label="tanh")
ax.plot(x, dx, color="#1abd15", linewidth=3, label="đạo hàm")
ax.legend(loc="upper left", frameon=False) #Hiển thị chú thích cho đồ thị
fig.show()
```

### VD5: Vẽ đồ thị hàm sigmoid

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1/(1+np.e**(-x))

x = np.arange(-8, 8, 0.01)
y = sigmoid(x)
dx = y*(1-y)

fig, ax = plt.subplots(figsize=(10,5))
ax.spines['left'].set_position('center') #chuyển trục bên trái về trung tâm figure
ax.spines['right'].set_color('none') #xóa trục phải
ax.spines['top'].set_color('none') #xóa trục trên
ax.xaxis.set_ticks_position('bottom') #thiết lập trục x là trục dưới
ax.yaxis.set_ticks_position('left') #thiết lập trục y là trục trái

ax.plot(x, y, color="#d35400", linewidth=3, label="sigmoid")
ax.plot(x, dx, color="#1abd15", linewidth=3, label="đạo hàm")
ax.legend(loc="upper left", frameon=False) #Hiển thị chú thích cho đồ thị
fig.show()
```

### VD6: Vẽ đồ thị hàm ReLU

```python
import matplotlib.pyplot as plt
import numpy as np

def relu_func(data):
    return (np.abs(data) + data) / 2
def relu_derivative(data):
    return (data > 0) * 1

data = np.arange(-8,8,0.01)
data_relu = relu_func(data)
data_drelu = relu_derivative(data)

# Chỉnh vị trí các trụ đồ thị
fig, ax = plt.subplots(figsize=(10,5))
ax.spines['left'].set_position('center') #chuyển trục bên trái về trung tâm figure
ax.spines['right'].set_color('none') #xóa trục phải
ax.spines['top'].set_color('none') #xóa trục trên
ax.xaxis.set_ticks_position('bottom') #thiết lập trục x là trục dưới
ax.yaxis.set_ticks_position('left') #thiết lập trục y là trục trái

ax.plot(data, data_relu, color="#d35400", linewidth=3, label="relu")
ax.plot(data, data_drelu, color="#1abd15", linewidth=3, label="đạo hàm")
ax.legend(loc="upper left", frameon=False) #Hiển thị chú thích cho đồ thị
fig.show()
```
